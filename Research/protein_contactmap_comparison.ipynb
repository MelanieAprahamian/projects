```python
import numpy as np
import math
import glob

pdb_list = ['1fgy','1jfm','1tpm','1w1h','2klx','2kr9','2mx1','2n28','2nc2','2y4q','3iql','3kq0','3uen','3vdj','4k47','4omo','4u9v','5lcs','5um3','6ao3']

for i, pdb in enumerate(pdb_list):
    
    with open('inputs/' + pdb + '.fasta','r') as fasta:
        for line in fasta:
            if not line.startswith('>'):
                num_aa = len(line.strip('\n'))
    
    #construct native contact map
    contact_map_native = np.zeros(shape=(num_aa,num_aa))
    
    native_residues = []
    with open('inputs/' + pdb + '.pdb','r') as native:
        for line in native:
            if len(line.split()) > 2:
                if (line.split()[2] == 'CB' and line.split()[3] != 'GLY') or (line.split()[2] == 'CA' and line.split()[3] == 'GLY'):
                    coords = {}
                    coords = {
                        'resid': line.split()[5],
                        'x': float(line.split()[6]),
                        'y': float(line.split()[7]),
                        'z': float(line.split()[8])
                    }
                    native_residues.append(coords)

    for i, res1 in enumerate(native_residues):
        for j, res2 in enumerate(native_residues):
            dist = math.sqrt((res2['x']-res1['x'])**2 + (res2['y']-res1['y'])**2 + (res2['z']-res1['z'])**2)
            if dist <= 8.0:
                contact_map_native[i,j] = 1
    
    #construct contact maps for top models for each residue type set
    for test in ['vanilla','all','common','optimal']:

        contact_map_model = np.zeros(shape=(num_aa,num_aa))
    
        model_residues = []
        file_name = glob.glob('top_models/' + test + '/*' + pdb + '*.pdb')
        with open(file_name[0],'r') as model:
            for line in model:
                if len(line.split()) > 2:
                    if (line.split()[2] == 'CB' and line.split()[3] != 'GLY') or (line.split()[2] == 'CA' and line.split()[3] == 'GLY'):
                        coords = {}
                        coords = {
                            'resid': line.split()[5],
                            'x': float(line.split()[6]),
                            'y': float(line.split()[7]),
                            'z': float(line.split()[8])
                        }
                        model_residues.append(coords)

        for i, res1 in enumerate(model_residues):
            for j, res2 in enumerate(model_residues):
                dist = math.sqrt((res2['x']-res1['x'])**2 + (res2['y']-res1['y'])**2 + (res2['z']-res1['z'])**2)
                if dist <= 8.0:
                    contact_map_model[i,j] = 1
                
        print(test, pdb, np.sum(contact_map_native == contact_map_model)/np.size(contact_map_native))
    print('')
```
